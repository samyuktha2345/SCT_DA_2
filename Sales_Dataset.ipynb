```python
import pandas as pd
import numpy as np

np.random.seed(1)

rows = 300

data = {
    "Order ID": [f"ORD-{1000+i}" for i in range(rows)],
    "Order Date": pd.date_range(start="2019-01-01", periods=rows, freq="D"),
    "Customer Name": np.random.choice(
        ["Arun", "Meena", "Ravi", "Anita", "Rahul", "Priya", "Kiran", "Sneha"], rows
    ),
    "Region": np.random.choice(
        ["North", "South", "East", "West"], rows
    ),
    "Category": np.random.choice(
        ["Furniture", "Technology", "Office Supplies"], rows
    ),
    "Sales": np.random.choice([None, 1200, 2500, 3700, 4800, 6200], rows),
    "Profit": np.random.randint(-1000, 3000, rows)
}

df = pd.DataFrame(data)

# Add duplicate rows intentionally
df = pd.concat([df, df.iloc[:20]], ignore_index=True)

# Save to Excel
df.to_excel("Large_Sales_Dataset.xlsx", index=False)

df.head()
```
